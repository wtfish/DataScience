Prprocessing with z-score and min-max

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../Datasets/iris_missing.csv")
```

```python
df.head()
```

a) detect null value, replace with median for each row

```python
kondisi1 = df['sepal_length'] == 0.0
kondisi2 = df['sepal_width'] == 0.0
kondisi3 = df['petal_length'] == 0.0
kondisi4 = df['petal_width'] == 0.0
df.loc[kondisi1 | kondisi2 | kondisi3 | kondisi4]
```

b) replace with np.NaN so zeroes can detected as null by isnull()

```python
df_null=df.replace(0.0, np.NaN)

kondisi1 = df_null['sepal_length'].isnull()
kondisi2 = df_null['sepal_width'].isnull()
kondisi3 = df_null['petal_length'].isnull()
kondisi4 = df_null['petal_width'].isnull()
df_null.loc[kondisi1 | kondisi2 | kondisi3 | kondisi4]
```

c) create imputation function that transform null values into it's class mean

```python
def imputation(df_input):
  list_columns = df_input.columns
  class_column = list_columns[-1]
  for column in list_columns[:-1]:
    df_input[column] = df_input[column].fillna(df_input.groupby(class_column)[column].transform('mean'))
  return df_input
```

```python
df_clean=imputation(df_null)
```

```python
df_clean.isnull().values.any()
```

```python
df_clean.head()
```

```python
def get_range(df_input):
  features = df_input.columns[:-1]
  for feature in features:
    max = df_input[feature].max()
    min = df_input[feature].min()
    print("range of",feature, " is ", max-min)
```

```python
get_range(df_clean)
```

Normalisasi min-max

```python
def minmax(df):
  features = df.columns[:-1]
  for feature in features:
    max = df[feature].max()
    min = df[feature].min()
    df[feature] = (df[feature]-min) / (max - min)
  return df
```

```python
df_minmaxnormal = minmax(df_clean)
df_minmaxnormal
```

```python
get_range(df_minmaxnormal)
```
